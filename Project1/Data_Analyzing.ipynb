# Check Text Column

```python
text_cols = [
    'Product',
    'PaymentMethod',
    'OrderStatus',
    'CouponCode',
    'ReferralSource'
]

for col in text_cols:
    print(df[col].unique())
```

## Clean Text

```python
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```
