```python
import pandas as pd

# 1. Dubara data load karein
df = pd.read_excel(
    "E:/project1_of_decodelab/Dataset for Data Analytics.xlsx"
)

# 2. CouponCode ki missing values ko handle karein (Project 1 ka task)
df['CouponCode'] = df['CouponCode'].fillna("No Coupon")

# 3. TotalPrice wali financial anomaly ko completely fix karein
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# 4. AB ASLI KAAM: Is clean data ko aik NAYI file mein permanently save karein
df.to_csv("final_cleaned_data.csv", index=False)

print("Mubarak ho! Asli data ab professionally clean ho kar 'final_cleaned_data.csv' mein save ho gaya hai.")
```
