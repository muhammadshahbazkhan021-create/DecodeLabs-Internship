Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Dataset

```python
df=pd.read_csv("E:\project1_of_decodelab\cleaned_data.csv")
```

Creating Visualization

Revenue By Product

```python
df.groupby('Product')['TotalPrice'].sum().plot(
    kind='bar'
)

plt.title('Revenue By Product')
plt.show()
```

Monthly Revenue Trend

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].dtype
# Month extract karo
df['Month'] = df['Date'].dt.to_period('M')

# Monthly revenue
monthly_sales = df.groupby('Month')['TotalPrice'].sum()

# Plot
monthly_sales.plot(
    kind='line',
    figsize=(10,5),
    title='Monthly Sales Trend'
)
```

Payment method

```python
df['PaymentMethod'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
```

Order Status

```python
df['OrderStatus'].value_counts().plot(
    kind='bar'
)
```

TotalPrice Distribution

```python
df['TotalPrice'].hist()
```

Outlier Detection

```python
sns.boxplot(x=df['TotalPrice'])
```
