### Detecting Outliers

```python

def check_outliers(column):
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    

    print(f"--- Outliers Analysis for {column} ---")
    print(f"Normal Range: {lower_bound:.2f} to {upper_bound:.2f}")
    print(f"Total Outliers Detected: {len(outliers)}\n")
    
    return outliers


price_outliers = check_outliers('TotalPrice')
qty_outliers = check_outliers('Quantity')

```
