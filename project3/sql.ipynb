# Importing Libraries 

```python
import pandas as pd
import sqlite3
```

### Loading Dataset

```python
df = pd.read_csv("cleaned_data.csv")
```

### Creating Database

```python
conn = sqlite3.connect("ecommerce.db")
```

```python
df.to_sql(
    "orders",
    conn,
    if_exists="replace",
    index=False
)
```

### Checking Database working properly

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

### Total Orders

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM orders
"""

run_query(query)
```

### Total Revenue

```python
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders
"""

run_query(query)
```

### Average Order Value

```python
query = """
SELECT AVG(TotalPrice) AS Avg_Order_Value
FROM orders
"""

run_query(query)
```

### Checking Higher Product Revenue

```python
query = """
SELECT Product,
SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY Product
ORDER BY Revenue DESC
"""

run_query(query)
```

### Checking Total Orders by Order Status

```python
query = """
SELECT OrderStatus,
COUNT(*) AS Total_Orders
FROM orders
GROUP BY OrderStatus
"""
run_query(query)
```

### Checking Revenue by Payment Method

```python
query = """
SELECT PaymentMethod,
SUM(TotalPrice) AS Revenue
FROM orders
GROUP BY PaymentMethod
ORDER BY Revenue DESC
"""

run_query(query)
```

### Checking Refferal Source

```python
query = """
SELECT ReferralSource,
COUNT(*) AS Orders
FROM orders
GROUP BY ReferralSource
ORDER BY Orders DESC
"""

run_query(query)
```

### Checking the orders that's price greater than 3000

```python
query = """
SELECT Count(*)
FROM orders
WHERE TotalPrice > 3000
ORDER BY TotalPrice DESC
"""

run_query(query)
```

### SQL Concepts Applied

* SELECT
* WHERE
* ORDER BY
* GROUP BY
* COUNT()
* SUM()
* AVG()
